```python
import numpy as np
from sklearn.cluster import KMeans

# Features: [Sprint_Speed, Stamina]
# Note: There is NO y variable here!
X_scaled = np.array([
    [95, 40], [92, 45], [98, 35], # High speed, low stamina
    [40, 95], [45, 92], [35, 98]  # Low speed, high stamina
])

# TODO 1: Initialize the KMeans model. Set n_clusters to 2, and random_state to 42.
kmeans_model = KMeans(n_clusters=2, random_state=42)

# TODO 2: Use .fit_predict() on the data to get the assigned clusters.
# Save the output to a variable called 'assigned_clusters'
assigned_clusters = kmeans_model.fit_predict(X_scaled)

# Let's print out the players and their new AI-assigned tactical group
for i in range(len(X_scaled)):
    print(f"Player {i+1} Stats: {X_scaled[i]} --> Assigned to Group: {assigned_clusters[i]}")
```

```python
import numpy as np
from sklearn.decomposition import PCA

# 5 Players, 5 Stats each: [Pace, Shoot, Pass, Dribble, Defend]
X_5D = np.array([
    [90, 85, 70, 88, 30],
    [88, 90, 65, 85, 35],
    [40, 50, 85, 70, 92],
    [45, 40, 88, 75, 90],
    [70, 75, 80, 80, 75]
])

# TODO 1: Initialize the PCA model and set n_components to 2
pca_model = PCA(n_components=2)

# TODO 2: Fit and transform the 5D data into 2D data
# Save it to a variable called 'X_2D'
X_2D = pca_model.fit_transform(X_5D)

# Let's verify the shape changed from (5, 5) to (5, 2)
print(f"Original Shape: {X_5D.shape}")
print(f"New Squashed Shape: {X_2D.shape}")
print(f"First player's new 2D coordinates: {np.round(X_2D[0], 2)}")
```
