```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans

# Scouting Data: [Crosses per 90, Tackles Won per 90, Sprint Speed (km/h)]
data = {
    'Player': ['Trent', 'Wan-Bissaka', 'Hakimi', 'Robertson', 'Walker', 'James', 'Zinchenko'],
    'Crosses': [8.5, 1.2, 5.0, 6.5, 2.0, 7.0, 3.5],
    'Tackles': [1.5, 4.5, 2.0, 2.5, 3.5, 2.8, 2.0],
    'Speed': [31.5, 33.0, 36.4, 32.0, 37.3, 33.5, 30.0]
}

df = pd.DataFrame(data)
features = df[['Crosses', 'Tackles', 'Speed']]

# TODO 1: Initialize the KMeans algorithm to find 3 tactical profiles.
# Ensure you set a random_state (e.g., 42) for reproducibility so our scouting report is consistent.
kmeans_model = KMeans(n_clusters=3, random_state=42)

# TODO 2: Fit the model on the feature data and predict the cluster labels in one step.
df['Tactical_Role_Cluster'] = kmeans_model.fit_predict(features)

# View the assignments
print(df[['Player', 'Tactical_Role_Cluster']])
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# Generating synthetic Liverpool academy data (Pressing & Passing)
# We secretly know there are 4 natural profiles, but the algorithm doesn't!
features, _ = make_blobs(n_samples=200, centers=4, cluster_std=1.5, random_state=42)
df = pd.DataFrame(features, columns=["Pressing_Intensity", "Passing_Accuracy"])

wcss_values = []
k_range = range(1, 11)

# TODO 1: Write a for-loop iterating through k_range.
# In each iteration, initialize KMeans with n_clusters=k.
# Set random_state=42 and n_init=10 to avoid warnings.
# Fit the model to our 'df', then append the model's 'inertia_' to the wcss_values list.

# YOUR CODE HERE
for k in k_range:
    kmeans_model = KMeans(n_clusters=k, random_state=42)
    kmeans_model.fit(df)
    wcss_values.append(kmeans_model.inertia_)

# View the calculated WCSS values
for k, wcss in zip(k_range, wcss_values):
    print(f"Roles (K) = {k}: WCSS = {wcss:.2f}")
```

```python
import pandas as pd
from sklearn.cluster import DBSCAN
import numpy as np

# Spatial Pass Data: [Pitch X-Coordinate, Pitch Y-Coordinate]
# Pitch is 105x68 meters.
data = {
    "Pass_ID": [1, 2, 3, 4, 5, 6, 7],
    "Pitch_X": [80.5, 81.0, 79.5, 20.0, 85.0, 84.5, 50.0],
    "Pitch_Y": [60.0, 61.5, 59.0, 10.0, 20.0, 21.0, 34.0]
}

df = pd.DataFrame(data)
pitch_coordinates = df[["Pitch_X", "Pitch_Y"]]

# TODO 1: Initialize DBSCAN.
# We want to group passes that are within 3.0 meters of each other (eps=3.0).
# We require at least 2 passes to form a recognized passing hub (min_samples=2).
dbscan_model = DBSCAN(eps=3.0, min_samples=2)

# TODO 2: Fit the model and extract the cluster labels.
# Remember, a label of -1 indicates Noise!
df["Passing_Hub"] = dbscan_model.fit_predict(pitch_coordinates)

print(df)
```

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Scouting Database: 5 Dimensions
# [Crosses, Tackles, Interceptions, Key Passes, Dribbles]
data = {
    "Player": ["Alexander-Arnold", "Robertson", "White", "Zinchenko", "Trippier", "Walker"],
    "Crosses": [8.5, 6.0, 2.5, 3.0, 7.5, 1.5],
    "Tackles": [1.5, 2.5, 3.0, 2.0, 2.5, 3.5],
    "Interceptions": [1.0, 1.5, 2.5, 1.8, 1.5, 2.0],
    "Key_Passes": [2.8, 1.5, 0.8, 1.2, 2.5, 0.5],
    "Dribbles": [1.2, 1.0, 0.5, 2.0, 0.8, 1.5]
}

df = pd.DataFrame(data)
features = df[["Crosses", "Tackles", "Interceptions", "Key_Passes", "Dribbles"]]

# CRITICAL STEP: PCA requires scaled data.
# We must standardize stats so a stat measured in hundreds (like passes)
# doesn't overpower a stat measured in decimals (like xG).
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# TODO 1: Initialize PCA to compress the data down to exactly 2 principal components.
pca_model = PCA(n_components=2)

# TODO 2: Fit the PCA model to the scaled_features and transform the data.
# This will output a new array with only 2 columns (PC1 and PC2).
principal_components = pca_model.fit_transform(scaled_features)

# Add the new "Super Stats" back to the dataframe
df["PC1"] = principal_components[:, 0]
df["PC2"] = principal_components[:, 1]

print(df[["Player", "PC1", "PC2"]])
```

```python
import numpy as np

# Mean-centered scouting data (Rows = Players, Cols = Stats)
# Stat 1: Pass Completion, Stat 2: Vision
X = np.array([
    [ 2.5,  2.4],
    [ 0.5,  0.7],
    [ 2.2,  2.9],
    [ 1.9,  2.2],
    [ 3.1,  3.0],
    [ 2.3,  2.7],
    [ 2.0,  1.6],
    [ 1.0,  1.1],
    [ 1.5,  1.6],
    [ 1.1,  0.9]
])

# To mean-center it perfectly (just in case):
X_centered = X - np.mean(X, axis=0)

# TODO 1: Calculate the Covariance Matrix using the mathematical formula.
# Hint: You need to transpose X_centered and multiply it by X_centered.
# In numpy, use .T for transpose and @ for matrix multiplication.
# Don't forget to divide by (n - 1) where n is the number of players.
n_players = X_centered.shape[0]
covariance_matrix = (1 / (n_players -1) ) * X_centered.T @ X_centered

# TODO 2: Calculate the Eigenvalues and Eigenvectors of the covariance_matrix.
# Hint: Use np.linalg.eig()
eigenvalues, eigenvectors = np.linalg.eig(covariance_matrix)

print("Covariance Matrix:\n", covariance_matrix)
print("\nEigenvalues:\n", eigenvalues)
print("\nEigenvectors (Principal Components):\n", eigenvectors)
```

```python
import pandas as pd
from sklearn.manifold import TSNE
from sklearn.datasets import make_moons

# Simulating complex, non-linear pressing data (e.g., intertwining tactical profiles)
features, tactical_labels = make_moons(n_samples=300, noise=0.05, random_state=42)

# TODO 1: Initialize the TSNE model.
# Set n_components to 2.
# Set perplexity to 30.
# Set random_state to 42 for reproducibility.
tsne_model = TSNE(n_components=2, perplexity=30, random_state=42)

# TODO 2: Fit the TSNE model and transform the features into 2D space.
# Note: Unlike PCA, t-SNE does NOT have a simple .transform() method for new data.
# You must use .fit_transform() directly on the dataset you want to map.
tsne_components = tsne_model.fit_transform(features)

df = pd.DataFrame()
df["tSNE_Dimension_1"] = tsne_components[:, 0]
df["tSNE_Dimension_2"] = tsne_components[:, 1]
df["Original_Tactical_Label"] = tactical_labels

print(df.head())
```

```python
import pandas as pd
import umap
from sklearn.datasets import make_moons

# Simulating complex, non-linear pressing data
features, tactical_labels = make_moons(n_samples=300, noise=0.05, random_state=42)

# TODO 1: Initialize the UMAP reducer.
# Set n_neighbors to 15 (a balanced view of local and global).
# Set min_dist to 0.1 (tight, but readable clusters).
# Set random_state to 42.
umap_reducer = umap.UMAP(random_state=42, n_neighbors=15, min_dist=0.1)

# TODO 2: Fit the UMAP model and transform the features into 2D space.
umap_components = umap_reducer.fit_transform(features)

df = pd.DataFrame()
df["UMAP_Dimension_1"] = umap_components[:, 0]
df["UMAP_Dimension_2"] = umap_components[:, 1]
df["Original_Tactical_Label"] = tactical_labels

print(df.head())
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

# MOCK SCOUTING DATABASE: 10 Attacking Metrics
np.random.seed(42)
data = np.random.rand(500, 10) * 100 # 500 Strikers, 10 metrics
feature_names = [f"Attacking_Metric_{i}" for i in range(1, 11)]
df_scouting = pd.DataFrame(data, columns=feature_names)

# Step 1: Standardization (Already completed for you)
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df_scouting[feature_names])

# TODO 1: Dimensionality Reduction
# Initialize PCA to keep exactly 3 Principal Components.
# Fit and transform the 'scaled_features' to create your compressed dataset.
pca_model = PCA(n_components=3)
compressed_pitch = pca_model.fit_transform(scaled_features)

# TODO 2: Tactical Profiling
# Initialize K-Means to find 4 distinct striker profiles (n_clusters=4).
# Set random_state to 42 for consistency.
kmeans_model = KMeans(n_clusters=4, random_state=42)

# TODO 3: The Pipeline Execution
# Fit the K-Means model ON THE 'compressed_pitch' (the PCA output).
# Predict the cluster labels and assign them to a new column in our dataframe.
df_scouting['Striker_Profile'] = kmeans_model.fit_predict(compressed_pitch)

# View the final Director of Football report (first 5 players)
print(df_scouting[['Attacking_Metric_1', 'Striker_Profile']].head())
```
