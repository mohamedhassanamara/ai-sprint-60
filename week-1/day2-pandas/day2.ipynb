```python
import pandas as pd
import numpy as np

# THE RAW, MESSY DATA
raw_data = {
    "Match_ID": [101, 101, 102, 103, 103, 104],
    "Player": ["Vinicius Jr", "Bellingham", "Rashford", "Fernandes", "Mainoo", "Rodrygo"],
    "Club": ["Real Madrid", "Real Madrid", "Man Utd", "Man Utd", "Man Utd", "Real Madrid"],
    "Role": ["Attacker", "Midfielder", "Attacker", "Midfielder", "Midfielder", "Attacker"],
    "Distance_km": [10.2, 11.8, np.nan, 12.1, 11.5, 9.8], # Rashford's tracker broke
    "Passes_Attempted": [40, 65, 35, 80, 55, np.nan],     # Rodrygo's pass data missing
    "Passes_Completed": [32, 58, 25, 60, 49, 38]
}

df = pd.DataFrame(raw_data)

# --- CAPSTONE PIPELINE ---

# TODO 1: Handle the Chaos.
# Rashford's distance is missing. Fill any NaN in "Distance_km" with the MEAN of the "Distance_km" column.
df["Distance_km"] = df["Distance_km"].fillna(df["Distance_km"].mean())
# Rodrygo's passes attempted is missing. We will use amputation here. Drop any rows that STILL contain a NaN value.
df = df.dropna(axis=0)


# TODO 2: Feature Engineering.
# Create a new column called "Pass_Accuracy" (Passes_Completed divided by Passes_Attempted).
df["Pass_Accuracy"] = df["Passes_Completed"] / df["Passes_Attempted"]

# TODO 3: Tactical Slicing.
# Create a new DataFrame called 'midfielders_df' that ONLY contains players where the "Role" is "Midfielder".
midfielders_df = df[df["Role"] == "Midfielder"]


# TODO 4: Grouping & Aggregation.
# Using your new 'midfielders_df', group by "Club" and calculate:
# - The mean of "Distance_km"
# - The mean of your newly engineered "Pass_Accuracy"
# Save this final matrix to a variable called 'final_tactical_summary'.
final_tactical_summary = midfielders_df.groupby("Club").agg({
    "Distance_km": 'mean',
    "Pass_Accuracy": 'mean',
})
```
